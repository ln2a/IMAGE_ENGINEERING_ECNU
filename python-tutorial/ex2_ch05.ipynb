```python
class TinyDetector(nn.Module):    def __init__(self):    super().__init__()    self.features = nn.Sequential(    nn.Conv2d(3,16,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),    nn.Conv2d(16,32,3,padding=1), nn.ReLU(), nn.MaxPool2d(2))    self.shared = nn.Sequential(nn.Flatten(),    nn.Linear(32*16*16,128), nn.ReLU())    self.cls_head = nn.Linear(128, 2) # 类别    self.box_head = nn.Linear(128, 4) # 边界框def forward(self, x):    h = self.shared(self.features(x))    return self.cls_head(h), torch.sigmoid(self.box_head(h))
```

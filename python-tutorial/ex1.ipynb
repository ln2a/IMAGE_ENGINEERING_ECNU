```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 直接读取同级目录下的 PNG 图片（OpenCV 默认 BGR 顺序）
img_bgr = cv2.imread('ex_img/lena_color_512.png')

# 转换为 RGB 顺序以便用 matplotlib 正确显示
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

def plot_res(images, titles):
    """在一行中显示一组图片。"""
    n = len(images)
    fig, axes = plt.subplots(1, n, figsize=(16, 8))

    if n == 1:
        axes = [axes]

    for ax, img, title in zip(axes, images, titles):
        ax.imshow(img)
        ax.set_title(title, fontsize=12)
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# 左右对比显示 BGR 和 RGB 的效果
plot_res(
    [img_bgr, img_rgb],
    ["OpenCV BGR (incorrect colors)", "Converted RGB (correct colors)"]
)
```
